```python
import numpy as np
import pandas as pd
import copy
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
print(tips.head())
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.lmplot(x="total_bill", y="tip", data=tips, line_kws={"color": "C1"})
plt.savefig('scatter-tips.png')
```
